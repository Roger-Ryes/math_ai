## Funcion compuesta

```python
import numpy as np
import matplotlib.pyplot as plt

N=1000
x = np.linspace(-10,10, N)

def g(x):
    return x**2

def f(x):
    return np.sin(x)

y = g(x)
f_o_g= f(y)
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(10,3))
ax1.plot(x,y)
ax1.set_title('funcion g(x)')
ax2.plot(x,f_o_g)
ax2.set_title('funcion f(g(x))')
```
